```python
df = pd.read_csv("Most Popular Programming Languages from 2004 to 2022.csv")
df["Date"] = pd.to_datetime(df["Date"])

df_popular = pd.DataFrame({"Languages":df.tail(1).T.index, "Popularity":df.tail(1).T.values.flatten()})
```

**VISUALISASI DATA**

Popularitas Proramming Language (January 2022)

```python
fig = px.pie(df_popular, values='Popularity', names='Languages',title='Programming Languages Score (January 2022)')
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
fig = px.bar(df_popular, x='Languages',y='Popularity',title='Programming Languages Score (January 2022)')
fig.show()
```

Terlihat berdasarkan visualisasi data, per Januari 2022, bahasa paling populer adalah Python dengan margin yang cukup besar jika dibandingkan dengan Java di posisi kedua. Dari grafik di atas, hal yang bisa kita peroleh adalah sebagai berikut:
i. Bahasa yang paling *worth it* untuk dicoba saat ini adalah Python.
ii. Tentu popularitas tersebut dilatarbelakingi oleh suatu alasan, terutama adalah faktor lapangan pekerjaan. Oleh karena itu, python merupakan programming language dengan lapangan pekerjaan terbanyak.
iii. Akan tetapi, data diats tidak menunjukkan detail sepunuhnya, karena diatas hanya menunjukkan popularitas dalam suatu waktu/momen (Januari 2022), dari grafik diatas saja tidak akan cukup untuk mengetahui keadaan suatu bahasa yang sedang menuju popularitas atau menuruni dalam popularitasnya.

Extra: Warna yang dipilih biru karena salah satu warna dari logo Python i.e. Biru dan Kuning

Popularitas Programming Language Berdasarkan Waktu

```python

df.plot(kind="line", x="Date", y=["C/C++","C#","Java","JavaScript","Kotlin","PHP","Python","Rust","Swift"],figsize=(20, 10))
plt.title("Popularity of languages")
plt.xlabel("Year")
plt.legend(bbox_to_anchor=(1,1), loc="upper left")
```

```python

df.plot(kind="bar", x="Date", y=["C/C++","C#","Java","JavaScript","Kotlin","PHP","Python","Rust","Swift"],figsize=(20, 10), stacked="True")
plt.title("Popularity of languages")
plt.xlabel("Year")
plt.legend(bbox_to_anchor=(1,1), loc="upper left")
```

Dua visualisasi diatas menunjukkan popularitas bahasa pemrograman berdasarkan waktu. Sekarang terlihat lebih jelas bahasa mana yang lebih layak untuk dipelajari. Hal yang bisa diperoleh:
i. Python tidak hanya populer saja, tetapi juga salah satu bahasa dengan index rising popularity yang tinggi.
ii. PHP walaupun merupakan salah satu bahasa pemrograman paling populer, ternyata index rising popularity nya rendah bahkan sudah memasuki fase menurun dari tahun ke tahun. Oleh karena itu, bagi yang ingin mempelajari PHP harus berpikir ulang, karena popularitasnya yang menurun tidak menjamin apakah akan masih relevan di masa yang akan datang.
iii. Bahasa pemrograman yang tahan uji waktu walaupun tidak memiliki lonjakan popularitas yang tinggi adalah C/C++, C#, JavaScript. Terlihat dari grafik di atas bahasa-bahasa tersebut cukup stabil untuk dipelajari. Selain itu, ke tiga bahasa tersebut juga bahasa-bahasa yang populer. (Terlihat kontras dibanding PHP yang hanya populer saat ini, tetapi index popularitasnya menurun.)

**DATA CLEANSING**

```python
df.isna().sum()
```

Penjelasan:
Fungsi .isna().sum() dibagi menjadi 2 tahap. 
Tahap 1 .isna() mencari data yang kosong dari tiap kolom.
Tahap 2 .sum() mengembalikan jumlah data yang kosong dari tiap kolom
Karena angka menunjukkan angka 0 semua maka tidak ada data yang kosong, alhasil tidak ada data yang perlu dibersihkan.
